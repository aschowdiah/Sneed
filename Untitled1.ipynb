```python
# Using the read_csv function and Path module, create a DataFrame 
unrefined08_df = pd.read_csv(
    Path("Resources/kmeans/2008.csv")
)

# Review the first and last five rows of the DataFrame
unrefined08_df 
```

```python
unrefined08_df.columns
```

```python
# Use drop function to eliminate non relevant data
sneed08_df = unrefined08_df[['Field', 'Transect', 'Trans sample','JohnGr 2008','LiBl 2008','BiBl 2008','Ind 2008','Swit 2008','EaGa 2008']]

# Review the first and last five rows of the DataFrame
sneed08_df
```

```python
sneed08_df= pd.get_dummies(sneed08_df,
                                 dummy_na=False, 
                                 columns=['Field','Transect','Trans sample'])
sneed08_df
```

```python
sneed08_df = sneed08_df.dropna()
sneed08_df
```

```python
sneed08_df
```

```python
sneed08_df = sneed08_df[sneed08_df['JohnGr 2008']!="ws"]
sneed08_df = sneed08_df[sneed08_df['JohnGr 2008']!="br"]
```

```python
sneed08_df.columns
```

```python
sneed08_df[['JohnGr 2008', 'LiBl 2008', 'BiBl 2008', 'Ind 2008', 'Swit 2008', 'EaGa 2008']] = sneed08_df[['JohnGr 2008', 'LiBl 2008', 'BiBl 2008', 'Ind 2008', 'Swit 2008',
       'EaGa 2008']].astype(int)
```

```python
sneed08_df.dtypes
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Create a list with the number of k-values to try
# Use a range from

k = list(range(1,22))
```

```python
# Create an empy list to store the inertia values
inertia = []
```

```python
# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `sneed04_clean_df`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    model = KMeans(n_clusters=i, random_state=0)
    model.fit(sneed08_df)
    inertia.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": k,
    "inertia": inertia}
```

```python
# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot = df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve 1", color='red', xticks=k)
elbow_plot
```

```python
model = KMeans(n_clusters=5, random_state=0)
model.fit(sneed08_df)
```

```python
prediction = model.predict(sneed08_df)
```

```python
sneed08_df['2008Prediction'] = prediction  
```
